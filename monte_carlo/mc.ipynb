```python
import numpy as np
import matplotlib.pyplot as plt
from timeit import default_timer as timer
from ising import spinMH,gpu_flag
# if you have nividia GPU and cupy installed, you can uncomment the following to have GPU support
#import cupy as cp
#from ising_gpu import spinMH,gpu_flag
```

```python
# setup
size = [1000,1000]
iters = 100
# run the function
start = timer()
state_ini,state_fin = spinMH(size=size,iters=iters)
end = timer()
if gpu_flag:
    print('GPU runtime in %.2f seconds'%(end-start))
else:
    print('CPU runtime in %.2f seconds'%(end-start))
```

```python
# To plot the results from the GPU version, please uncomment the following two
#state_ini = cp.asnumpy(state_ini)
#state_fin = cp.asnumpy(state_fin)

fig , ax = plt.subplots()
plt.subplot(121)
plt.imshow(state_ini,interpolation=None,cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('initial')
plt.subplot(122)
plt.imshow(state_fin,interpolation=None,cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title(str(iters)+' iterations')
plt.show()
```
