```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rndgen(seed=314159,size=None):
    a = 2**18+1
    c = 7
    if size is None:
        return (a*seed + 7)%2**32
    else:
        rnd_ls = [seed]
        i = 1
        while i <= size:
            u_i = (a*rnd_ls[-1] + 7)%2**32
            rnd_ls.append(u_i)
            i += 1
        return np.asarray(rnd_ls)
```

```python
ui = rndgen(size=1000)
```

```python
plt.scatter(ui[:-1]/2**32,ui[1:]/2**32)
```

```python
plt.hist(ui/2**32)
```
